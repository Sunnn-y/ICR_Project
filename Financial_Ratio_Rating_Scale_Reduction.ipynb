```python
import pandas as pd
import re

data = pd.read_csv('/content/drive/MyDrive/ICR/재무비율(건설업계)_백테스트/건설업계_재무비율_백테스트_신용등급_merge.csv')

def reductuion(text):
    return ''.join(re.findall('[a-zA-Z]', str(text))) # 정규표현식을 사용하여 영어만 추출하는 함수 정의

data['등급'] = data['등급'].apply(reductuion) # apply 함수를 사용하여 열에 적용
```

```python
data.to_csv('/content/drive/MyDrive/ICR/재무비율(건설업계)_백테스트/건설업계_재무비율_백테스트_신용등급_merge_등급단계축소.csv')
```
