```python
# 무한대 값을 임의의 큰값으로 대체
import math
df.replace(float(math.inf), float(1.5), inplace=True)
```

```python
## 정규화

from sklearn.preprocessing import MinMaxScaler

minmax_scaler = MinMaxScaler()

columns_to_normalize = df[['부채비율', '차입금의존도', '고정장기적합률', '자기자본순이익률', '매출액영업이익률', '매출액경상이익률', '이자보상배율(배)',
                                        '매출액 증가율', '총자산 증가율', '영업이익 증가율', '자기자본 증가율', '매출채권 회전기일(일)', '재고자산 회전기일(일)', '매입채무 회전기일(일)',
                                        '총자산 회전율', '부채상환계수(배)','총차입금상환능력비율']]

minmax_scaled = minmax_scaler.fit_transform(columns_to_normalize)

df[['부채비율', '차입금의존도', '고정장기적합률', '자기자본순이익률', '매출액영업이익률', '매출액경상이익률', '이자보상배율(배)',
                 '매출액 증가율', '총자산 증가율', '영업이익 증가율', '자기자본 증가율', '매출채권 회전기일(일)', '재고자산 회전기일(일)', '매입채무 회전기일(일)',
                 '총자산 회전율', '부채상환계수(배)','총차입금상환능력비율']] = minmax_scaled

pd.DataFrame(df)
df
```

```python
# 안정성 점수 계산
weights = [0.25, 0.25, 0.25, 0.25]   # 가중치

def calculate_composite_index(row):
    return weights[0] * (1 - row['부채비율']) + weights[1] * row['자기자본순이익률'] + weights[2] * (1 - row['차입금의존도']) + weights[3] * row['고정장기적합률']

df['안정성'] = df.apply(calculate_composite_index, axis=1)

df
```

```python
# 수익성 점수 계산
weights = [0.34, 0.33, 0.33]  # 가중치

def calculate_composite_index(row):
    return weights[0] * row['매출액영업이익률'] + weights[1] * row['매출액경상이익률'] + weights[2] * row['이자보상배율(배)']

df['수익성'] = df.apply(calculate_composite_index, axis=1)

df
```

```python
# 성장성 점수 계산
weights = [0.25, 0.25, 0.25, 0.25]   # 가중치

def calculate_composite_index(row):
    return weights[0] * row['매출액 증가율'] + weights[1] * row['총자산 증가율'] + weights[2] * row['영업이익 증가율'] + weights[3] * row['자기자본 증가율']

df['성장성'] = df.apply(calculate_composite_index, axis=1)

df
```

```python
# 활동성 점수 계산
weights = [0.25, 0.25, 0.25, 0.25]    # 가중치

def calculate_composite_index(row):
    return weights[0] * (1 - row['매출채권 회전기일(일)']) + weights[1] * (1 - row['재고자산 회전기일(일)']) + weights[2] * (1 - row['매입채무 회전기일(일)']) + weights[3] * row['총자산 회전율']

df['활동성'] = df.apply(calculate_composite_index, axis=1)

df
```

```python
# 현금흐름 점수 계산
weights = [0.5, 0.5]    # 가중치

def calculate_composite_index(row):
    return weights[0] * (1 - row['부채상환계수(배)']) + weights[1] * (1 - row['총차입금상환능력비율'])

df['현금흐름'] = df.apply(calculate_composite_index, axis=1)

df
```

```python
## 카테고리별 평균 점수 계산
selected_columns = ['차입금의존도', '고정장기적합률', '매출액경상이익률', '매출액 증가율', '영업이익 증가율',
                    '매출채권 회전기일(일)', '재고자산 회전기일(일)', '부채상환계수(배)', '부채비율', '자기자본순이익률',
                    '매출액영업이익률', '이자보상배율(배)', '총자산 증가율', '자기자본 증가율', '매입채무 회전기일(일)',
                    '총자산 회전율', '총차입금상환능력비율', '안정성', '수익성', '활동성', '성장성', '현금흐름']

average_values = df[selected_columns].mean()

df.loc['평균'] = pd.Series({'이름': '평균', **average_values})
```

```python
df = pd.DataFrame(df)

df
```

```python
## 총점 계산
df['총점'] = df[['안정성', '수익성', '활동성', '성장성', '현금흐름']].sum(axis=1)
df
```
