 # 백테스트할 데이터 concat

```python
import os
import glob
import pandas as pd

def concat_all_csv_files(folder_path, output_path=None):

    # 폴더 내 모든 CSV 파일 경로 수집
    file_paths = glob.glob(os.path.join(folder_path, '*.csv'))

    if not file_paths:
        raise ValueError("No CSV files found in the specified folder.")

    # 불러오기 및 합치기
    result_df = pd.concat(map(pd.read_csv, file_paths), ignore_index=True)

    # 결과를 출력 파일로 저장
    if output_path:
        result_df.to_csv(output_path, index=False)

    return result_df

# 사용 예시
folder_path = "/content/drive/MyDrive/ICR/재무비율(건설업계)_백테스트_최종"
output_path = "/content/drive/MyDrive/ICR/재무비율(건설업계)_백테스트_최종/건설업계_재무비율_백테스트_최종_.csv"
result_df = concat_all_csv_files(folder_path, output_path)
```
